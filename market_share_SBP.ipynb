```python
import pandas as pd
import os
```

```python
os.chdir(r'C:\1 Benchmarking')
os.getcwd()
#os.listdir()
```

```python
contenido = os.listdir()
bancos = []
for banco in contenido:
    if os.listdir() and banco.startswith('BANCO_en_'):
        bancos.append(banco)
print(bancos)
```

```python
data = pd.DataFrame()
for banco in bancos:
    data1 = pd.read_excel(banco, skiprows= 9)
    data1['banco'] = banco.replace("BANCO_en_", "").replace(".xlsx", "")  
    data = data.append(data1,ignore_index=True)
```
